```python
import os

# ✅ Directly set your real dataset path (BEST METHOD)
dataset_path = "C:/Users/ASUS/Desktop/Dataset/train"

print(" Using dataset path:", dataset_path)

# Verify folders exist
print("Images:", os.listdir(os.path.join(dataset_path, "images"))[:5])
print("Labels:", os.listdir(os.path.join(dataset_path, "labels"))[:5])
```

```python
import os

data_yaml = f"""
path: {os.path.abspath(dataset_path)}
train: images
val: images

names:
  0: scratch
  1: edge_damage
"""

with open("data.yaml", "w") as f:
    f.write(data_yaml)

print(" data.yaml created!")

# Check file
with open("data.yaml", "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8s.pt")

# Train
model.train(
    data="data.yaml",
    epochs=30,
    imgsz=640,
    batch=8
)
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")

print(" Model loaded successfully!")
```

```python
import os

os.makedirs("outputs", exist_ok=True)

print(" Output folder ready!")
```

```python
import cv2
import json
import os

image_dir = os.path.join(dataset_path, "images")
output_dir = "outputs"

for img_name in os.listdir(image_dir):
    img_path = os.path.join(image_dir, img_name)

    results = model(img_path)[0]
    image = cv2.imread(img_path)

    detections = {}

    if results.boxes is not None:
        for i, box in enumerate(results.boxes.xyxy):
            x_min, y_min, x_max, y_max = map(int, box.tolist())

            # Draw box
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

            detections[str(i+1)] = {
                "bbox_coordinates": [x_min, y_min, x_max, y_max]
            }

    # Save image
    cv2.imwrite(os.path.join(output_dir, img_name), image)

    # Save JSON
    json_path = os.path.join(output_dir, img_name.rsplit(".", 1)[0] + ".json")
    with open(json_path, "w") as f:
        json.dump(detections, f, indent=4)

print(" Inference complete!")
```
